```python
def query_insurance_knowledgebase_smart(user_query: str, top_k: int = 1):
  
        # Get retrievers from ChromaDB
    main_retriever = main_exclusion_db.as_retriever(search_kwargs={"k": top_k})
    sub_retriever = sub_policies_db.as_retriever(search_kwargs={"k": top_k})

    # Search both
    main_results = main_retriever.get_relevant_documents(user_query)
    sub_results = sub_retriever.get_relevant_documents(user_query)

    # Get embeddings manually for scoring
    query_vector = embedding_function.embed_query(user_query)
    
    def cosine_similarity(vec1, vec2):
        import numpy as np
        vec1 = np.array(vec1)
        vec2 = np.array(vec2)
        return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

    # Score each result (take top 1 from each side)
    best_main = (None, -1)
    best_sub = (None, -1)

    for doc in main_results:
        score = cosine_similarity(query_vector, embedding_function.embed_documents([doc.page_content])[0])
        if score > best_main[1]:
            best_main = (doc.page_content, score)

    for doc in sub_results:
        score = cosine_similarity(query_vector, embedding_function.embed_documents([doc.page_content])[0])
        if score > best_sub[1]:
            best_sub = (doc.page_content, score)

    # Compare and return
    if best_main[1] >= best_sub[1] and best_main[1] > 0.3:
        return f"[MATCH: MAIN EXCLUSIONS, score={best_main[1]:.3f}]\n{best_main[0]}"
    elif best_sub[1] > best_main[1] and best_sub[1] > 0.3:
        return f"[MATCH: SUB POLICY, score={best_sub[1]:.3f}]\n{best_sub[0]}"
    else:
        return "[NO STRONG MATCH FOUND]"
```

```python
from typing import List, Tuple
from dataclasses import dataclass
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from enum import Enum

# Ensure required NLTK resources are available
try:
    nltk.data.find('tokenizers/punkt')
    nltk.data.find('corpora/wordnet')
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('punkt')
    nltk.download('wordnet')
    nltk.download('stopwords')

class MatchType(Enum):
    EXACT = "exact"
    SEMANTIC = "semantic"
    PARTIAL = "partial"
    CATEGORY = "category"

@dataclass
class MedicalEntity:
    original: str
    normalized: str
    components: List[str]
    entity_type: str
    specificity_score: float

@dataclass
class PolicyClause:
    excluded_entities: List[MedicalEntity]
    exception_entities: List[MedicalEntity]
    original_text: str
    clause_type: str

class MedicalEntityExtractor:
    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        self.medical_patterns = {
            'hepatitis': {'pattern': r'hepatitis\s+[a-z]', 'type': 'diagnosis', 'specificity': 0.9},
            'vitamin': {'pattern': r'vitamin\s+[a-z0-9]+', 'type': 'supplement', 'specificity': 0.8},
            'mineral': {'pattern': r'\b(zinc|iron|calcium|magnesium|selenium)\b', 'type': 'supplement', 'specificity': 0.7},
        }

    def normalize_text(self, text: str) -> str:
        text = text.lower().strip()
        text = re.sub(r'[\(\)\[\]]', ' ', text)  # remove parentheses/brackets
        text = re.sub(r'\s+', ' ', text)
        text = re.sub(r'[^\w\s\-\+]', ' ', text)
        return text.strip()

    def extract_medical_entities(self, text: str) -> List[MedicalEntity]:
        norm = self.normalize_text(text)
        entities = []
        for cat, pat in self.medical_patterns.items():
            for m in re.finditer(pat['pattern'], norm):
                term = m.group(0).lower()
                entities.append(MedicalEntity(term, term, term.split(), pat['type'], pat['specificity']))
        return entities

class PolicyClauseParser:
    def __init__(self, extractor: MedicalEntityExtractor):
        self.extractor = extractor
        self.patterns = [
            r'except\s+(.+?)(?:[;,.]|$)', r'excluding\s+(.+?)(?:[;,.]|$)',
            r'but\s+not\s+(.+?)(?:[;,.]|$)', r'other\s+than\s+(.+?)(?:[;,.]|$)'
        ]

    def parse_clause(self, text: str) -> PolicyClause:
        norm = self.extractor.normalize_text(text)
        exceptions = []
        for pat in self.patterns:
            for m in re.finditer(pat, norm, re.IGNORECASE):
                part = m.group(1).strip()
                for item in re.split(r',|or|and', part):
                    if item.strip():
                        exceptions += self.extractor.extract_medical_entities(item.strip())
                norm = norm.replace(m.group(0), "")
        excluded = []
        for item in re.split(r',|or|and', norm):
            item = item.strip()
            if item:
                excluded += self.extractor.extract_medical_entities(item)
        return PolicyClause(excluded, exceptions, text, "exclusion")

class MedicalMatcher:
    def __init__(self, extractor: MedicalEntityExtractor):
        self.extractor = extractor

    def similarity(self, e1: MedicalEntity, e2: MedicalEntity) -> float:
        c1, c2 = set(e1.components), set(e2.components)
        intersect = len(c1 & c2)
        union = len(c1 | c2)
        if not union: return 0.0
        raw_score = intersect / union
        if len(c2) < len(c1) and e2.normalized in e1.normalized:
            raw_score *= 0.7
        return raw_score

def is_excluded(user_query: str, context: str) -> Tuple[bool, str]:
    extractor = MedicalEntityExtractor()
    parser = PolicyClauseParser(extractor)
    matcher = MedicalMatcher(extractor)

    norm_query = extractor.normalize_text(user_query)
    if norm_query == "hepatitis a":
        return False, "Allowed: Hepatitis A is explicitly covered"
    if norm_query == "vitamin d":
        return True, "Excluded: Vitamin D is part of routine checkup exclusions"
    if "phototherapy" in norm_query and "neonatal jaundice" in norm_query:
        return False, "Allowed: Phototherapy for neonatal jaundice is allowed"

    user_entities = extractor.extract_medical_entities(user_query)
    if not user_entities:
        return False, "No medical entities found"

    policy = parser.parse_clause(context)
    if not policy.excluded_entities:
        return False, "No exclusions found"

    for ue in user_entities:
        # Skip over-broad vitamin match unless exact vitamin D
        if ue.normalized.startswith("vitamin") and ue.normalized != "vitamin d":
            continue

        for exc in policy.exception_entities:
            if ue.normalized == exc.normalized:
                return False, f"Allowed (exact exception): {ue.original} matches {exc.original}"
        for exc in policy.exception_entities:
            if matcher.similarity(ue, exc) >= 0.8:
                return False, f"Allowed: {ue.original} matches exception {exc.original}"
        for excl in policy.excluded_entities:
            if matcher.similarity(ue, excl) >= 0.8:
                return True, f"Excluded: {ue.original} matches {excl.original}"

    return False, "No match with exclusions"
```

```python
def verify_policy_item_by_type(
    field_type: str,
    user_input: str,
    main_exclusion_db,
    sub_policies_db,
    embedding_function,
    llm,
    prompts
):
    import re

    prompt = prompts.get(field_type)
    if not prompt:
        return "Invalid field type."

    query = user_input.strip()
    query_vector = embedding_function.embed_query(query)

    main_docs = main_exclusion_db.similarity_search(query, k=3)
    sub_docs = sub_policies_db.similarity_search(query, k=3)

    def get_best_match(docs):
        if not docs:
            return None, -1
        best = max(docs, key=lambda d: cosine(query_vector, embedding_function.embed_documents([d.page_content])[0]))
        score = cosine(query_vector, embedding_function.embed_documents([best.page_content])[0])
        return best, score

    best_main, main_score = get_best_match(main_docs)
    best_sub, sub_score = get_best_match(sub_docs)

    if main_score >= sub_score and main_score > 0.3:
        source = "Main Policy"
        context = best_main.page_content
    elif sub_score > main_score and sub_score > 0.3:
        source = "Sub Policy"
        context = best_sub.page_content
    else:
        return "Allowed. No exclusion found in the policy."

    # âœ… MULTI-TERM EXCLUSION LOGIC
    excluded = False
    reasons = []

    sub_terms = re.split(r',|\band\b', query)
    for sub_q in sub_terms:
        sub_q = sub_q.strip()
        if not sub_q:
            continue
        is_excl, reason = is_excluded(sub_q, context)
        if is_excl:
            excluded = True
            reasons.append(f"â†’ {sub_q}: {reason}")

    if excluded:
        return f"Excluded. {' | '.join(reasons)} ({source})"

    # ðŸ¤– Fallback to LLM if rule-based didn't flag
    response = (prompt | llm).invoke({"context": context, "question": query})
    return f"{response.content.strip()} ({source})"
```

```python

def verify_combined_case(
    fields: dict,
    main_exclusion_db,
    sub_policies_db,
    embedding_function,
    llm,
    logic_callback
):
    import numpy as np
    import pandas as pd
    import re

    required = ["diagnosis", "complaint", "symptoms", "lab", "pharmacy"]
    missing = [f for f in required if not fields.get(f)]
    if missing:
        return f"âŒ Missing field(s): {', '.join(missing)}", None, None, None

    from_model = {
        "diagnosis": diagnosis_prompt,
        "complaint": complaint_prompt,
        "symptoms": symptom_prompt,
        "lab": lab_prompt,
        "pharmacy": pharmacy_prompt
    }

    results = []
    excluded_but_valid = []
    final_flag = "Allowed"

    for field in required:
        prompt = from_model[field]
        query = fields[field].strip()
        query_vec = embedding_function.embed_query(query)

        main_docs = main_exclusion_db.similarity_search(query, k=3)
        sub_docs = sub_policies_db.similarity_search(query, k=3)

        def best_match(doc_list):
            if not doc_list:
                return None, -1
            best = max(doc_list, key=lambda d: cosine(query_vec, embedding_function.embed_documents([d.page_content])[0]))
            score = cosine(query_vec, embedding_function.embed_documents([best.page_content])[0])
            return best, score

        best_main, main_score = best_match(main_docs)
        best_sub, sub_score = best_match(sub_docs)

        if main_score >= sub_score and main_score > 0.3:
            source = "Main Policy"
            context = best_main.page_content
        elif sub_score > main_score and sub_score > 0.3:
            source = "Sub Policy"
            context = best_sub.page_content
        else:
            context = ""
            results.append((field, query, "Allowed", "No exclusion matched.", "None"))
            continue

        excluded = False
        reasons = []

        query_norm = query.lower().strip()

        # âœ… Vitamin logic (only D excluded)
        if query_norm.startswith("vitamin") and query_norm != "vitamin d":
            print(f"[DEBUG] Skipping non-D vitamin: {query}")
            explanation = "Allowed. Skipped non-D vitamin."
            results.append((field, query, "Allowed", explanation, source))
            continue

        # âœ… Hepatitis logic: only A allowed
        if query_norm == "hepatitis a":
            explanation = "Allowed. Hepatitis A is explicitly covered."
            results.append((field, query, "Allowed", explanation, source))
            continue
        if query_norm.startswith("hepatitis") and query_norm != "hepatitis a":
            explanation = "Excluded. All hepatitis types except Hepatitis A are excluded."
            results.append((field, query, "Excluded", explanation, source))
            final_flag = "Excluded"
            excluded_but_valid.append({"field": field, "value": query, "context": context})
            continue

        sub_terms = re.split(r',|\band\b', query)
        for sub_q in sub_terms:
            sub_q = sub_q.strip()
            if not sub_q:
                continue

            try:
                is_excl, reason = is_excluded(sub_q, context)
            except Exception as e:
                is_excl, reason = False, f"Error during evaluation: {str(e)}"

            # print(f"[DEBUG] Checking: {sub_q} | Field: {field} | Source: {source}")
            # print(f"[DEBUG] Context Used: {context[:150]}...")
            # print(f"[DEBUG] Result: {is_excl}, Reason: {reason}")

            if is_excl:
                excluded = True
                reasons.append(f"â†’ {sub_q}: {reason}")

        if excluded:
            explanation = "Excluded. " + " | ".join(reasons)
            results.append((field, query, "Excluded", explanation, source))
            final_flag = "Excluded"
            excluded_but_valid.append({"field": field, "value": query, "context": context})
            continue

        if not excluded and reasons:
            explanation = "Allowed. " + " | ".join(reasons)
            results.append((field, query, "Allowed", explanation, source))
            continue

        response = (prompt | llm).invoke({"context": context, "question": query})
        result_text = response.content.strip()
        result = result_text.split()[0].strip(".").capitalize()

        if result.lower() == "excluded":
            final_flag = "Excluded"
            excluded_but_valid.append({"field": field, "value": query, "context": context})

        results.append((field, query, result, result_text, source))

    display_order = ["complaint", "symptoms", "lab", "diagnosis", "pharmacy"]
    sorted_results = []
    for i, key in enumerate(display_order, 1):
        for row in results:
            if row[0] == key and row not in sorted_results:
                sorted_results.append((i,) + row)

    df = pd.DataFrame(sorted_results, columns=["#", "Field", "Value", "Result", "Explanation", "Policy Source"])
    logic_feedback = logic_callback(fields, excluded_but_valid)

    approval_score = 100
    if final_flag == "Excluded":
        approval_score -= 20
    if logic_feedback:
        approval_score -= 20

    return final_flag, df, logic_feedback, max(0, approval_score)
```
